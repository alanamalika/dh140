This function calculates how many hours and years it will take for someone to become a billionaire, without taxes or expense, if they earn w dollars per hour.

```python
# function (both a new function definition and an execution of that function)
def f(w):
    hours = 1000000000/w
    years = hours/8760
    saying = 'You will have to work ' + str(hours) + ' hours or ' + str(years) + ' years to become a billionaire if you earn ' + str(w) + ' dollars per hour before tax and expenses.'
    return (saying)
```
