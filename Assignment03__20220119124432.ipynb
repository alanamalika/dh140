```python
x = billionaires_df.drop(671)
x
```

```python
z = x.drop_duplicates(subset=['name'])
z
```

```python
m = z.groupby('location.region')[['name']].count()
m
```

```python
bar = m.plot(kind = 'barh')
bar.set_title('Regional Distribution of Billionaires', fontsize = 16)
bar.legend('Number of billionaires',fontsize = 10)
bar.set_xlabel('Number of billionaires',fontsize = 14)
bar.set_ylabel('Region',fontsize = 14)
```

```python
pie = m.plot.pie(y='name', figsize=(12, 10))
pie.set_title('Regional Distribution of Billionaires', fontsize = 16)
pie.set_xlabel('Number of billionaires',fontsize = 14)
pie.set_ylabel('Region',fontsize = 14)
```
